# Reading TSVs
When reading tsvs, use the read_csv() to read the tsv and csv files.
usecols is parameter similar to query. The parameter will select certain columns to be viewed in the data frame.

```python
df = pd.read_csv('files/simpsons-episodes.tsv',
                 sep = "\t",
                 header = None,
                 names=col_names,
                 skiprows=4,
                 index_col='Production code',
                 parse_dates=['Air date'],
                 na_values='no_val',
                 #usecols=select_cols,
                 encoding = 'utf8')
df
```
